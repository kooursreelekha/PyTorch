```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
s=StandardScaler()
X_train=s.fit_transform(X_train)
X_test=s.transform(X_test)
```

```python
#converting  numpy arrays to tenors
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
## defining model
class MyNN:
    def __init__(self,X):

        self.weights=torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)#here we need this for differentiation so requires is true
        self.bias=torch.zeros(1,dtype=torch.float64,requires_grad=True)#here we need this for differentiation so requires is true
    
    def forward_pass(self,X):
        z=torch.matmul(X,self.weights)+self.bias
        y_pred=torch.sigmoid(z)
        return y_pred
    #loss calculate
    def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
        loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
        return loss

    
    #backward pass
    #parameters update
```

```python
#parameters
learning_rate=0.1
epochs=25
```

```python
# create model
model=MyNN(X_train_tensor)
model.weights
```

```python
model.bias
```

```python
# training pipeline
model=MyNN(X_train_tensor)
#forward pass
for epoch in range(epochs):
    y_pred=model.forward_pass(X_train_tensor)
    # print(y_pred)
    # print(y_pred.shape)
    
# loss calculation
    loss=model.loss_function(y_pred,y_train_tensor)
    
    
# backward pass
    loss.backward()

# parameter update
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad

# zero gradients---for every epoch they are started with o
    model.weights.grad.zero_()
    model.bias.grad.zero_()
    
#loss in every epoch
    print(f'epoch:{epoch+1} ,loss:{loss.item()}')
```

```python
# evaluation
with torch.no_grad():
  y_pred = model.forward_pass(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
