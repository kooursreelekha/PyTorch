```python
import torch
import torch.nn as nn
```

```python
class Module(nn.Module):
    
    def __init__(self,num_features):
        
        super().__init__()   # calling constructor of nn.Module                               #
        self.linear=nn.Linear(num_features,1) # this takes no.of feat and no.of output        #  # --- we create layers in this region
        self.sigmoid=nn.Sigmoid()                                                             #

    def forward(self,features):
        
        out=self.linear(features) # here z=wx+b is calculated                                 #  
        out=self.sigmoid(out) # z is passed to sigmoid fun                                    #  # ---- here we use them
        return out                                                                            #
        
        
        
```

```python
# create dataset
features=torch.rand(10,5)

#create model
model=Module(features.shape[1])

#call model forward pass
model(features) # can also do model.forward(features)
```
